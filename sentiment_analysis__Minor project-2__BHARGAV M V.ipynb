```python
import pandas as pd
df= pd.read_csv("/content/drive/MyDrive/Musical_instruments_reviews.csv")
df
```

```python
df1=df.copy()
```

```python
df1['reviewText']=df1['reviewText'].fillna('Missing')
```

```python
df1['reviews']=df1['reviewText']+df1['summary']
```

```python
df1
 
```

```python
# 0 if review is negative, 1 is review is neutral 2 if review is positive
df1['sentiment']=0
for i in df1.index:
  if df1['overall'][i]==3.0:
    df1['sentiment'][i]=1
  elif df1['overall'][i]==1.0 or df1['overall'][i]==2.0:
    df1['sentiment'][i]=0
  else:
    df1['sentiment'][i]=2
```

```python
df1['sentiment'].value_counts()
```

```python
df1=df1.drop(['reviewerName','unixReviewTime'], axis=1)
```

```python
import re
import string
def review_cleaning(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
df1['reviews']=df1['reviews'].apply(lambda x:review_cleaning(x))
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
from spacy.lang.en import English
punctuations = string.punctuation
parser = English()
stopwords = list(STOP_WORDS)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
reviews=[]
for i in range(0, len(df1)):
  review = re.sub('[^a-zA-Z]', ' ', df1['reviews'][i])
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords]
  review = ' '.join(review)
  reviews.append(review)
labels = df1['sentiment'].values
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(reviews,labels, test_size=0.2, random_state=1000)
```

```python
def tokenizer(u):
      tokens = parser(u)
      return [token.lemma_.lower().strip() for token in tokens if token.text.lower().strip() not in stopwords and token.text not in punctuations]
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(tokenizer = tokenizer, ngram_range=(1,1))
vectorizer.fit(xtrain)
```

```python
xtrain = vectorizer.transform(xtrain)
xtest1 = vectorizer.transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
accuracy = model.score(xtest1, ytest)
print('Accuracy:', accuracy)
```
