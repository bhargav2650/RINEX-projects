```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
df=pd.read_csv("fruits.txt",sep="\t")
df.shape
```

```python
x = df.iloc[:,3:7].values
y = df.iloc[:,1:2].values
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state=42)
```

```python
print(xtrain.shape)
print(xtest.shape)
```

```python
#compare to min-max scaler z score gives more accuracy 
from scipy import stats
xtrain=stats.zscore(xtrain)
xtest=stats.zscore(xtest)
```

```python
#DecisionTree
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=4)
model.fit(xtrain,ytrain.ravel())
ypredd = model.predict(xtest)
confusion_matrix(ytest,ypredd)
accuracy_score(ypredd,ytest)*100
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(30,10))
tree.plot_tree(model,class_names=['apple','lemon','orange','mandarin'])
plt.show()
```

```python
#Naive Bayes
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
ypred = gnb.fit(xtrain, ytrain.ravel()).predict(xtest)
confusion_matrix(ypred,ytest)
accuracy_score(ypred,ytest)*100
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3,p=2)
model.fit(xtrain,ytrain.ravel())
ypred_knn = model.predict(xtest)
confusion_matrix(ypred_knn,ytest)
accuracy_score(ypred_knn,ytest)*100
```

```python
#SVM
from sklearn.svm import SVC
model =SVC(kernel='rbf')
model.fit(xtrain,ytrain.ravel())
ypred_svm = model.predict(xtest)
confusion_matrix(ypred_svm,ytest)
accuracy_score(ypred_svm,ytest)*100
```

```python
#Random forest
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(xtrain,ytrain.ravel()) 
y_pred_r = model.predict(xtest)
confusion_matrix(y_pred_r,ytest)
accuracy_score(y_pred_r,ytest)*100
```
